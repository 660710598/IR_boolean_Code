```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#download
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('punkt')

documents = [
    "This is a good party and everyone is having fun.",
    "The party is not over yet, we are still dancing!",
    "It was an excellent party with wild music and great friends.",
    "The party is over, but we had an amazing time!",
    "Wild party happened last night."
]
```

```python
def boolean_search(documents, query):
    stop_words = set(stopwords.words("english"))

    def preprocess(text):
        tokens = word_tokenize(text.lower())
        return [t for t in tokens if t.isalnum() and t not in stop_words]

    # แยก query
    terms = query.lower().split()
    include_terms = []
    exclude_terms = []

    skip = False
    for i, term in enumerate(terms):
        if skip:
            skip = False
            continue
        if term == "not":
            exclude_terms.append(terms[i + 1])
            skip = True
        elif term != "and":
            include_terms.append(term)

    results = []
    for doc in documents:
        tokens = preprocess(doc)

        # AND condition
        if not all(term in tokens for term in include_terms):
            continue

        # NOT condition
        if any(term in tokens for term in exclude_terms):
            continue

        results.append(doc)

    return results
```

```python
#Boolean Search
query = "good AND party NOT over"

matching_docs = boolean_search(documents, query)

print(f"ผลลัพธ์การค้นหาสำหรับคำค้น '{query}':")
for doc in matching_docs:
    print(f"- {doc}")
```
