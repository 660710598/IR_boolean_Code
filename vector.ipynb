```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ดาวน์โหลดข้อมูลที่จำเป็นสำหรับ NLTK (ถ้ายังไม่มี)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # สำหรับเวอร์ชันใหม่ๆ

# 1. สร้าง Corpus ใหม่ (เปลี่ยนจากวิทยาศาสตร์ เป็นเรื่องผลไม้และสุขภาพ)
corpus = [
    'Apple is a nutritious fruit high in fiber and vitamin C',
    'Oranges are citrus fruits known for immune system support and vitamin C',
    'Bananas are rich in potassium and great for energy boost',
    'Strawberries are sweet red berries full of antioxidants',
    'Regular exercise and a healthy diet prevent chronic diseases',
    'Running and swimming are good cardiovascular exercises for health',
    'Dark chocolate in moderation can provide antioxidants'
]
```

```python
print(stopwords.words('english'))
stop_words = nltk.corpus.stopwords.words('English')
```

```python
# ฟังก์ชันตัดคำ (Tokenization) [อ้างอิง: 30-32]
def get_tokenized_list(doc_text):
    tokens = nltk.word_tokenize(doc_text)
    return tokens

# ฟังก์ชันลดรูปคำ (Stemming) [อ้างอิง: 34-39]
def word_stemmer(token_list):
    ps = nltk.stem.PorterStemmer()
    stemmed = []
    for words in token_list:
        stemmed.append(ps.stem(words))
    return stemmed
```

```python
# ฟังก์ชันลบ Stopwords [อ้างอิง: 41-46]
def remove_stopwords(doc_text):
    cleaned_text = []
    for words in doc_text:
        if words.lower() not in stop_words: 
            cleaned_text.append(words)
    return cleaned_text
```

```python
doc_text = get_tokenized_list(corpus[2])
print("WORD TOKENS:")
print(doc_text)
doc_text = remove_stopwords(doc_text)
print("\nAFTER REMOVING STOPWORDS:")
print(doc_text)
print("\nAFTER PERFOMING THE WORD STEMMING::")
doc_text = word_stemmer(doc_text)
doc_text
```

```python
cleaned_corpus = []
for doc in corpus:
    tokens = get_tokenized_list(doc)
    doc_text = remove_stopwords(tokens)
    doc_text = word_stemmer(doc_text)
    doc_text = ' '.join(doc_text) 
    cleaned_corpus.append(doc_text)
cleaned_corpus
```

```python
doc_vector = vectorizer.transform(cleaned_corpus)
print(vectorizer.get_feature_names_out())
```

```python
df1 = pd.DataFrame(doc_vector.toarray(), columns=vectorizer.get_feature_names_out())
df1
```

```python
query = 'vitamin C and antioxidants fruits' 
query_tokens = get_tokenized_list(query)
query_tokens = remove_stopwords(query_tokens)
q = []
for w in word_stemmer(query):
    q.append(w)
q = ' '.join(q)
q
query_vector = vectorizer.transform([q])
df1 = pd.DataFrame(query_vector.toarray(), columns=vectorizer.get_feature_names_out())
df1
```

```python
cosineSimilarities = cosine_similarity(doc_vector, query_vector).flatten()
```

```python
related_docs_indices = cosineSimilarities.argsort()[:-10:-1]
print (related_docs_indices)

for i in related_docs_indices:
    data = [cleaned_corpus[i]]
    print(data)
```
