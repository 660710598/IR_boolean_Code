```python
from bs4 import BeautifulSoup
import requests
import csv

# Fetching the webpage
url = "https://cookpad.com/eng/search/japan"
res = requests.get(url)
res.encoding = "utf-8"
print(res)

# Checking the status code
if res.status_code == 200:
    print("Successful")
elif res.status_code == 404:
    print("Error 404 page not found")
else:
    print("Not both 200 and 404")

# Creating the BeautifulSoup object
soup = BeautifulSoup(res.text, 'html.parser')
# print(soup) 
# print(soup.prettify()) # Formats the HTML for easier reading

# Examples of accessing specific HTML tags (from Image 2)
# print(soup.title)
# print(soup.h1)
# print(soup.p)
# print(soup.title.string)

# Finding all elements of a specific tag (from Image 3)
# a_tags = soup.find_all('a')

# Extracting course titles from h2 tags
courses = soup.find_all('h2')
#print(courses)

course_list = []
for course in courses:
   title = course.get_text(strip=True)

   if title:  # Check if the title is not empty
       course_list.append(title)
       print(title)

csv_data = [["Recipe Title"]]
for title in course_list:
    csv_data.append([title])

# Saving the extracted data to a CSV file 
with open('Japan_Food.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(csv_data)

print("\n‡∏ö‡∏±‡∏ô‡∏ó‡∏∂‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏•‡∏á‡πÑ‡∏ü‡∏•‡πå Japan_Food.csv ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß‡∏Ñ‡∏£‡∏±‡∏ö!")
```
