```python
from bs4 import BeautifulSoup
import requests
import csv

# 1. Fetching the webpage
url = "http://stackpython.co/courses"
res = requests.get(url)
res.encoding = "utf-8"
print(res)

# 2. Checking the status code
if res.status_code == 200:
    print("Successful")
elif res.status_code == 404:
    print("Error 404 page not found")
else:
    print("Not both 200 and 404")

# 3. Creating the BeautifulSoup object
soup = BeautifulSoup(res.text, 'html.parser')
# print(soup) 
# print(soup.prettify()) # Formats the HTML for easier reading

# 4. Examples of accessing specific HTML tags (from Image 2)
# print(soup.title)
# print(soup.h1)
# print(soup.p)
# print(soup.title.string)

# 5. Finding all elements of a specific tag (from Image 3)
# a_tags = soup.find_all('a')

# 6. Extracting course titles from h2 tags
courses = soup.find_all('h2')
print(courses)

course_list = []
for course in courses:
    obj = course.string
    course_list.append(obj)

# 7. Saving the extracted data to a CSV file (from Image 4)
csv_col = [['title'], [course_list]]
f = open('course_titiles.csv', 'w') # Transcribed exactly as in the image
with f:
    writer = csv.writer(f)
    for row in csv_col:
        writer.writerow(row)
```
